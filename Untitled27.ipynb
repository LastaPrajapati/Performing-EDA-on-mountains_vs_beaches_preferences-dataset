Code for loading and displaying key information about the dataset

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
preferences_data = pd.read_csv('/content/mountains_vs_beaches_preferences.csv')

# Display basic information
print("Dataset Info:")
print(preferences_data.info())

# Describe the dataset for summary statistics
print("\nSummary Statistics:")
print(preferences_data.describe(include='all'))

# Check for missing values
print("\nMissing Values per Column:")
print(preferences_data.isnull().sum())
```

Code for generating a histogram for numerical data and a count plot for categorical data

```python
# Distribution of Age and Preference
plt.figure(figsize=(12, 5))

# Age Distribution
plt.subplot(1, 2, 1)
sns.histplot(preferences_data['Age'], kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Preference Distribution
plt.subplot(1, 2, 2)
sns.countplot(x=preferences_data['Preference'], palette='viridis')
plt.title('Preference Distribution (Mountains vs Beaches)')
plt.xlabel('Preference')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

Code for automatically detecting and plotting histograms for all numerical columns

```python
# Plot histograms for each numerical feature
numerical_features = preferences_data.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(15, 20))
for i, column in enumerate(numerical_features, 1):
    plt.subplot(6, 3, i)
    plt.hist(preferences_data[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'{column} Distribution')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Code for producing boxplots to show the distribution

```python
# Bivariate analysis with boxplots
selected_features = ['Age', 'Income', 'Vacation_Budget']

plt.figure(figsize=(15, 12))
for i, feature in enumerate(selected_features, 1):
    plt.subplot(3, 2, i)
    sns.boxplot(data=preferences_data, x='Preference', y=feature, palette='Set2')
    plt.title(f'{feature} vs Preference')

plt.tight_layout()
plt.show()
```

Code for generating a heatmap showing the correlation between each pair of numerical features

```python
# Correlation matrix to understand relationships between numerical features
plt.figure(figsize=(14, 10))

# Select only numerical features for correlation calculation
numerical_data = preferences_data.select_dtypes(include=['number'])
correlation_matrix = numerical_data.corr()  # Calculate correlation on numerical data

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title('Feature Correlation Matrix')
plt.show()
```

Code for Count Plot showing the distribution of genders and distribution of Favorite_Season

```python
# Gender and Favorite Season distribution
plt.figure(figsize=(12, 5))

# Gender Distribution
plt.subplot(1, 2, 1)
sns.countplot(x=preferences_data['Gender'], palette='pastel')
plt.title('Gender Distribution')

# Favorite Season Distribution
plt.subplot(1, 2, 2)
sns.countplot(x=preferences_data['Favorite_Season'], palette='Set3')
plt.title('Favorite Season Distribution')

plt.tight_layout()
plt.show()
```

Code for generating histograms with KDE plots

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of numerical features to visualize (update based on actual numerical columns in the dataset)
numerical_features = [
    'Age', 'Income', 'Vacation_Budget', 'Proximity_to_Mountains',
    'Proximity_to_Beaches'
]

# Setting up the plot grid
plt.figure(figsize=(15, 20))
for i, column in enumerate(numerical_features, 1):
    plt.subplot(4, 3, i)
    sns.histplot(preferences_data[column], kde=True, color='skyblue', edgecolor='black')
    plt.title(f'{column} Distribution')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Code for creating boxplots to highlight potential outliers in each feature

```python
# Boxplot to identify potential outliers in Income and Vacation Budget
outlier_features = ['Income', 'Vacation_Budget']

plt.figure(figsize=(10, 5))
for i, feature in enumerate(outlier_features, 1):
    plt.subplot(1, 2, i)
    sns.boxplot(x=preferences_data[feature], color='skyblue')
    plt.title(f'Outliers in {feature}')

plt.tight_layout()
plt.show()
```
